```python
from pathlib import Path
```

```python
raw_dir = Path("../data/raw")
json_files = sorted(raw_dir.glob("*.json"))
```

```python
import json

all_games = []
files_processed = 0

for file_path in json_files:
    with file_path.open("r", encoding="utf-8") as f:
        data = json.load(f)

    results = data.get("results", [])

    # opcional: validar tipo
    if not isinstance(results, list):
        continue

    all_games.extend(results)
    files_processed += 1

    print(f"{file_path.name} -> {len(results)} juegos")

print(f"Archivos procesados: {files_processed}")
print(f"Total juegos acumulados: {len(all_games)}")
```

```python
print(all_games[2])
```

```python
platforms_dict = {}

for game in all_games:
    for platform in game.get("platforms", []):
        platform_id = platform.get("id")
        platform_name = platform.get("name")
    platforms_dict[platform_id] = platform_name
```

```python
print(len(platforms_dict))
print(list(platforms_dict.items())[:10])
```

```python
genres_data = [
    {"genre_id": g, "name": n}
    for g, n in genres_dict.items()
]
```

```python
import pandas as pd
genres_df = pd.DataFrame(genres_data)
```
